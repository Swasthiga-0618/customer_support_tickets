```python
import pandas as pd

def get_samples_by_queue_priority(df, queue_name, priority_level, sample_size=10):
    """
    Gets a sample of tickets based on queue and priority.

    Args:
        df: The input DataFrame.
        queue_name: The name of the queue.
        priority_level: The priority level.
        sample_size: The number of samples to retrieve (default is 10).

    Returns:
        A list of sample tickets (as dictionaries), or an empty list if no matching tickets are found.
    """
    subset = df[(df["queue"] == queue_name) & (df["priority"] == priority_level)]
    if not subset.empty:
        samples = subset.sample(min(sample_size, len(subset))).to_dict(orient="records")
        return samples
    else:
        return []

# Example Usage (assuming your DataFrame is named 'df'):
# Updated queue names:
technical_support_low_priority = get_samples_by_queue_priority(df, "Technical Support", "low")
print("Technical Support, Low Priority:")
for sample in technical_support_low_priority:
    print(sample)

product_support_low_priority = get_samples_by_queue_priority(df, "Product Support", "low")
print("\nProduct Support, Low Priority:")
for sample in product_support_low_priority:
    print(sample)
```
