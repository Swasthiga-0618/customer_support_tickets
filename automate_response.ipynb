```python
# prompt: automate the issue responses

import pandas as pd

# Load the issue templates from the CSV file
try:
    template_data = pd.read_csv("issue_templates.csv")
except FileNotFoundError:
    print("Error: issue_templates.csv not found. Please run the code that creates it first.")
    exit()  # Exit the script if the file is not found

def create_response(issue_type, user_input=None):
    """
    Creates an automated response based on the issue type and optional user input.

    Args:
        issue_type: The type of issue reported.
        user_input: A dictionary of additional information provided by the user.

    Returns:
        A string containing the automated response.
    """
    template = template_data[template_data['Issue Type'] == issue_type]

    if template.empty:
      return f"No template found for issue type '{issue_type}'. Please provide more details."

    template = template.iloc[0].to_dict()
    response = f"Issue Type: {template['Issue Type']}\n"
    response += f"Steps to Follow: {template['Steps to Follow']}\n"

    if user_input:
        for key, value in user_input.items():
            response += f"{key}: {value}\n"

    return response

# Example usage:
issue_type = "Login Issue"  # Get this from user input or other source
user_details = {"User ID/Email": "testuser@example.com", "Description": "Unable to log in."}
automated_response = create_response(issue_type, user_details)
print(automated_response)

issue_type = "Payment Failure"
user_details = {"Transaction ID": "12345", "Description": "Payment declined."}
automated_response = create_response(issue_type, user_details)
print(automated_response)

issue_type = "Unknown Issue Type"
automated_response = create_response(issue_type)
automated_response
```
